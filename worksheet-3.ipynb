### Question 1.2
What does _compression_ of an image mean?
What is the difference between lossless and lossy compression?

### Answer 1.2
*Compression* of an image imply the byte minimizing of the graphical file without degrading the image quality to an unaccaptable level, and the file can be decompressed again to the large file. 

Lossless compression refers to compression in which the compressed images is pixel-for-pixel identical to the original image. The image is reduced without any quality loss. By lossy compression the image will lose data.

When a lossless compressed image is restored to its original form, no loss of data occurs. On the contrary, when a lossy compressed image is restored to its original form, data is lost.

## Task 3: vector fields and quiver plots

![Quiver](https://machinelearningknowledge.ai/wp-content/uploads/2020/09/10.2-Quiver.png)

A vector field is, as it says, mapping which to every point in some domain associates a vector. One can think of it in terms of a fluid in motion: at each point the fluid has a certain velocity. Another way of thinking of a vector field is as the right-hand side in a [first order system of ordinary differential equations](https://en.wikipedia.org/wiki/Ordinary_differential_equation#System_of_ODEs).

To plot vector fields in two dimensions in Matplotlib one can use the `quiver` command. First learn about the `quiver` command, for example by studying [this tutorial](https://pythonforundergradengineers.com/quiver-plot-with-matplotlib-and-jupyter-notebooks.html).

### Task 3.1

Study the differential equations for the [non-linear pendulum](https://en.wikipedia.org/wiki/Pendulum_(mechanics)).
Make a velocity field plot for the pendulum system written as a first order system

$$
    \theta' = v, \qquad v' = -\sin\theta.
$$
Use the domain $-\pi \leq \theta \leq \pi$ and $-3\leq v \leq 3$.
Put $\theta$ on the $x$-axis and $v$ on the $y$-axis.

Modify the plot, adding color to the arrows corresponding to the magnitude of the velocity vectors. Use your own colormap. (**Hint:** use the `C` input to the `quiver` function).

```python
import matplotlib

# functin that calculates the derivate
def f(Y):
    y1, y2 = Y
    return [y2, -np.sin(y1)]


theta = np.arange(-np.pi, np.pi, 0.3)
v = np.arange(-3, 3, 0.3)

T, V = np.meshgrid(theta, v)

t_dir, v_dir = np.zeros(T.shape), np.zeros(V.shape)

I, J = T.shape

for i in range(I):
    for j in range(J):
        x = T[i, j]
        y = V[i, j]
        yprime = f([x, y])
        t_dir[i,j] = yprime[0]
        v_dir[i,j] = yprime[1]

# magnitude of the arrows
mag = np.hypot(t_dir, v_dir)
q = plt.quiver(T, V, t_dir, v_dir, mag, cmap = cubehelix_palette)

plt.xlabel('Angle')
plt.ylabel('Velocity')
plt.xlim([-np.pi, np.pi])
plt.ylim([-3, 3])
```
